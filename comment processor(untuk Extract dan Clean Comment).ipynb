```python
import random
from googleapiclient.discovery import build

# Masukkan API Key Anda di sini
API_KEY = 'AIzaSyD5n2MfSSZBg9Z9iEGBsf4oK0U02l89s3w'

def get_random_gaming_videos(query="gaming indonesia", min_comments=1000, max_results=50):
    youtube = build('youtube', 'v3', developerKey=API_KEY)

    # 1. Cari video berdasarkan kata kunci dan lokasi Indonesia
    search_request = youtube.search().list(
        q=query,
        part="snippet",
        type="video",
        relevanceLanguage="id",
        regionCode="ID",
        maxResults=2000  # Ambil sampel awal lebih banyak untuk difilter
    )
    search_response = search_request.execute()

    video_list = []

    # 2. Iterasi hasil pencarian untuk cek statistik (jumlah komentar)
    for item in search_response['items']:
        video_id = item['id']['videoId']
        video_title = item['snippet']['title']
        
        # Ambil statistik video
        stats_request = youtube.videos().list(
            part="statistics",
            id=video_id
        )
        stats_response = stats_request.execute()
        
        stats = stats_response['items'][0]['statistics']
        comment_count = int(stats.get('commentCount', 0))

        # 3. Filter berdasarkan minimal komentar
        if comment_count >= min_comments:
            video_list.append({
                'title': video_title,
                'link': f"https://www.youtube.com/watch?v={video_id}",
                'comments': comment_count
            })

    # 4. Acak hasil akhir
    random.shuffle(video_list)
    return video_list[:max_results]

# Jalankan fungsi
if __name__ == "__main__":
    try:
        results = get_random_gaming_videos()
        print(f"--- Ditemukan {len(results)} video yang sesuai kriteria ---\n")
        for i, vid in enumerate(results, 1):
            print(f"{i}. {vid['title']}")
            print(f"   Link: {vid['link']}")
            print(f"   Komentar: {vid['comments']}\n")
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")
```

```python
import pandas as pd
import re

# 1. LOAD DATASET MENTAH
df = pd.read_csv('comments_multi_video.csv')

def advanced_preprocessing(text):
    if not isinstance(text, str):
        return ""
    
    # a. Case Folding: Mengubah menjadi huruf kecil 
    text = text.lower()
    
    # b. Menghapus URL/Link 
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    
    # c. Menghapus Mentions (@username) dan Hashtag 
    text = re.sub(r'@\w+|\#\w+', '', text)
    
    # d. Menghapus Emoji dan Karakter Khusus (kecuali tanda baca dasar) [cite: 117, 128]
    text = re.sub(r'[^\w\s.,!?]', '', text)
    
    # e. Normalisasi Spasi: Menghapus spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

# 2. EKSEKUSI PEMBERSIHAN
df['comment_text'] = df['comment_text'].apply(advanced_preprocessing)

# 3. PEMBERSIHAN LANJUTAN (VALIDITAS DATA)
# Menghapus komentar yang kosong setelah dibersihkan [cite: 54]
df = df[df['comment_text'] != ""]

# Menghapus duplikat untuk menghindari bias 
df = df.drop_duplicates(subset=['comment_text'])

# 4. PENYIAPAN KOLOM SESUAI SKEMA PELABELAN [cite: 102, 105]
# Memastikan kolom sesuai dengan format: {id, comment_text, aspect_class, sentiment_class}
df = df[['comment_text']]
df.insert(0, 'id', range(1, len(df) + 1))
df['aspect_class'] = ""    # Target: Gameplay, Graphics, Monetization, Creator/Content, Community, Other
df['sentiment_class'] = "" # Target: Positive, Negative, Neutral

# 5. SIMPAN DATASET SIAP LABEL
df.to_csv('comments_ready_for_labelling.csv', index=False)

print(f"Preprocessing Selesai!")
print(f"Jumlah data unik dan bersih: {len(df)}")
print(f"File disimpan sebagai: comments_ready_for_labelling.csv")
```
